```python
import pickle
import matplotlib.pyplot as plt
from scipy.stats.mstats import gmean
import seaborn as sns
from statistics import stdev
from math import log
import numpy as np
from scipy import stats
from statistics import mean
%matplotlib inline
```

```python
size = [8000,18000,32000,50000,72000,96000]
stan_data = ["stan_total_time_8000.p","stan_total_time_18000.p", "stan_total_time_32000.p", "stan_total_time_50000.p", \
             "stan_total_time_72000.p", "stan_total_time_96000.p"]
multi_2 = ["multi_total_time_2_8000_v.p", "multi_total_time_2_18000_v.p", "multi_total_time_2_32000_v.p",\
           "multi_total_time_2_50000_v.p","multi_total_time_2_72000_v.p", "multi_total_time_2_96000_v.p" ]
multi_4 = ["multi_total_time_4_8000_v.p", "multi_total_time_4_18000_v.p", "multi_total_time_4_32000_v.p",\
           "multi_total_time_4_50000_v.p","multi_total_time_4_72000_v.p", "multi_total_time_4_96000_v.p" ]
```

```python
def read_in(results):
    container =[]
    time_mean = []
    for res in results:  
        out = pickle.load(open(res,"rb"))
        out_app = list(map(lambda x : (x/60)/60, out))
        container.append((max(out_app)-min(out_app)))
        time_mean.append(mean(out_app)) 
    return time_mean, container
    
```

```python
stan_mean, stan_results = read_in(stan_data)
multi_2_mean, multi_2_results = read_in(multi_2)
multi_4_mean, multi_4_results = read_in(multi_4)
```

```python
plt.figure(figsize=(9,9))


plt.errorbar(size, stan_mean, yerr=stan_results, fmt = '--o', label = "1 Processor")
plt.errorbar(size, multi_2_mean, yerr=multi_2_results, fmt = "-D", label = "2 Processors")
plt.errorbar(size, multi_4_mean, yerr=multi_4_results, fmt = ':H', label = "4 Processors")
plt.legend()
plt.title("Population Size and Run Time",  fontweight = "bold", fontsize = 20)
plt.xlabel("Number of Agents", fontsize = 15,fontweight = 'bold')
plt.ylabel("Time Per Run (Hours)", fontsize = 15,fontweight = 'bold')
```
