```python
size = [8000,18000,32000,50000,72000,96000]
stan_data = ["stan_total_time_8000.p","stan_total_time_18000.p", "stan_total_time_32000.p", "stan_total_time_50000.p", \
             "stan_total_time_72000.p", "stan_total_time_96000.p"]
multi_2 = ["multi_total_time_2_8000_v.p", "multi_total_time_2_18000_v.p", "multi_total_time_2_32000_v.p",\
           "multi_total_time_2_50000_v.p","multi_total_time_2_72000_v.p", "multi_total_time_2_96000_v.p" ]
multi_4 = ["multi_total_time_4_8000_v.p", "multi_total_time_4_18000_v.p", "multi_total_time_4_32000_v.p",\
           "multi_total_time_4_50000_v.p","multi_total_time_4_72000_v.p", "multi_total_time_4_96000_v.p" ]
```

```python
def read_in(results):
    container =[]
    time_mean = []
    for res in results:  
        out = pickle.load(open(res,"rb"))
        out_app = list(map(lambda x : (x/60)/60, out))
        container.append((max(out_app)-min(out_app)))
        time_mean.append(mean(out_app)) 
    return time_mean, container
    
```

```python
stan_mean, stan_results = read_in(stan_data)
multi_2_mean, multi_2_results = read_in(multi_2)
multi_4_mean, multi_4_results = read_in(multi_4)
```

# Step Time

```python
#size = [8000,18000,32000,50000,72000,96000]
stan_step= ["stan_time_8000.p","stan_time_18000.p", "stan_time_32000.p", "stan_time_50000.p", \
             "stan_time_72000.p", "stan_time_96000.p"]
multi_2_step = ["multi_step_time_2_8000_v.p", "multi_step_time_2_18000_v.p", "multi_step_time_2_32000_v.p",\
           "multi_step_time_2_50000_v.p","multi_step_time_2_72000_v.p", "multi_step_time_2_96000_v.p" ]
multi_4_step = ["multi_step_time_4_8000_v.p", "multi_step_time_4_18000_v.p", "multi_step_time_4_32000_v.p",\
           "multi_step_time_4_50000_v.p","multi_step_time_4_72000_v.p", "multi_step_time_4_96000_v.p" ]
```

```python
def split_step(data, processors):
    corrected = []
    for each in data: 
        each_o = pickle.load(open(each,"rb"))
        each_d = []
        for run in each_o: 
            each_d.append(run/processors)
        corrected.append(each_d)
    return corrected
```

```python
step_2_c = split_step(multi_2_step, 2)
step_4_c = split_step(multi_4_step, 4)
step_1_c = split_step(stan_step, 1)
```

```python
def get_base(total, step):
    base =[]
    every = []
    for each in range(len(total)):
        each_o = pickle.load(open(total[each],"rb"))
        each_d = []
        for run in range(len(each_o)):
           each_d.append(each_o[run]-step[each][run])
        every.append(mean(step[each]))
        base.append(mean(each_d))
    return base, every
            
```

```python
step_s, step_s_all = get_base(multi_2, step_2_c)
step_4, step_4_all = get_base(multi_4, step_4_c)
step_1, step_1_all = get_base(stan_data, step_1_c)
```

```python
def totals(data):
    res = []
    for i in data: 
        res.append(i)
    return res
```

```python
totals_2 = totals(multi_2_mean)
totals_4 = totals(multi_4_mean)
totals_1 = totals(stan_mean)
```

```python
plt.figure(figsize=(5,9))
ax = plt.axes()
y_ticks = [2000/60**2, 4000/60**2, 6000/60**2, 8000/60**2]
ax.set_yticklabels =(y_ticks)
plt.plot(size, step_1_all, "bo--",label = "Management 1 Processor")
plt.plot(size, step_1, "bo:", label = "Step 1 Processor")
#plt.plot(size, stan_mean, label = "1 Processor")
plt.plot(size, step_s_all, "gD--", label = "Management 2 Processors")
plt.plot(size, step_s, "gD:", label = "Step 2 Processors")
#plt.plot(size, totals_2, "gD-", label = "Total 2 Processors")
plt.plot(size, step_4_all, "rH--",label = "Management 4 Processors")
plt.plot(size, step_4, "rH:", label = "Step 4 Processors")
#plt.plot(size, totals_4, "rH-", label = "Total 4 Processors")

#plt.plot(size, totals_1, "bo-", label = "Total 1 Processor")
plt.legend()
plt.title("Time Per Step and Management",  fontweight = "bold", fontsize = 20)
plt.xlabel("Number of Agents", fontsize = 15,fontweight = 'bold')
plt.ylabel("Time Per Run (seconds)", fontsize = 15,fontweight = 'bold');

```

```python
size_x = [x for x in range(100000, 2000000, 100000)]
```

```python
def make_line(x, slope):
    y = []
    for i in x: 
        
        y.append(((slope*i)/60)/60)
        
    return y
```

```python
proc_1 = make_line(size_x, proc_1_slp)
proc_2 = make_line(size_x, proc_2_slp)
proc_4 = make_line(size_x, proc_4_slp)
```
